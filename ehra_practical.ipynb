## Linear carcinogens
- Estimation of risk is based on 
- unit risk estimate for inhalation

## What is URE
- URE = upper bound excess lifetime risk
- URE results from continuous exposure to agent over lifetime at a concentration of 1 microg/cumm in air
- One lifetime = 70 years
- if URE = 1.5 * 10^-6 microg/cu.mm, THEN:
- No more than 1.5 EXCESS tumours are expected to develop 
- per 1000000 people if they were exposed ALL day, EVERY day
- for a LIFETIME to a concentration of 1 microgram/cu.mm of air
- Upper bound of a risk
- Actual risk may be lower

## What is CPS
- Upper bound of a 95% Confidence
- increased cancer risk from a lifetime oral ingestion
- proportion of a population affected per mg/Kg/day
- assume 70 kg body weight
- Extrapolate from the low dose region of the dose-response relationship
- Exposures corresponding to risks < 1 in 100

## IARC (WHO) Classification

## Preamble to monographs (pp 31)
- Sufficient evidence: causal association between agent and human cancer established
- Limited evidence: causal interpretation is credible but cannot rule out chance, bias, and confounding with confidence
- Inadequate evidence: poor quality, consistency, statistical precision, no data for humans, 
- Lack of evidence: high quality consistent studies showing No positive association

## Non-carcinogenic effect
- Dose thresholds
- Lowest dose / safety factors = Margin of safety
- Minimal risk levels
- MRL: daily exposure to a substance S
- that exposure will be without appreciable risk of adverse effects
- the exposure must be over a specified period
- 0-14 days for acute effects
- 15-364 days for intermediate effets
- >= 365 days for long term or chronic effects
- Uncertainty factors are 10, 1, or 3
- Use LOAEL

## Critical effect

- If the critical effect is prevented, then the health effect will not result

## Concepts of dose-response calculations

## Non-linear dose-response assessment
- Threshold hypothesis
- Organisms (read: humans) can tolerate a range of exposures starting from 0 to some value
- Threshold of toxicity is the dosage where the effects START to happen
- Focus on the most sensitive members of the community
- Aim to keep the exposure BELOW threshold value

## Non-linear association
- mode of action suggests toxicity has a threshold
- Below the threshold no toxicity will be manifest
- above the threshold, toxicity
- NOAEL: highest exposure at which NO biologically significant inreases in the severity of adverse effect
- if NOAEL is NOT possible, then LOAEL
- What is the lowest dose at which biologically meaningful significant toxic effect gets manifest

## Reference dose
- Oral or dermal
- f(NOAEL | LOAEL, UF)
- Rfd = NOAEL / UF
- If data comes from human studies, UF = 10
- If data comes from animals, then humans, UF = 10 * 10
- If instead of NOAEL, LOAELs used, add 10 fold
- Rfd = LOAEL / 10 * 10 * 10 OR LOAEL = 10 * 10
- Alternatively, Rfd = LOAEL / sqrt(10) for human studies
- OR, Rfd = LOAEL / sqrt(10) * sqrt(10) = LOAEL / 10
- Estimate of a DAILY oral exposure to human that is likely to be WITHOUT an appreciable risk of toxic effect over a lifetime 

## Reference dose
- Expressed in mg/Kg/day

## Linear dose-response association
- Toxicity does not have a threshold
- 0 threshold
- Do not use UFs
- From the origin, (0,0), draw a line to the (dose, response) point where toxic effects manifest
- The point where toxic effects manifest is called departure point
- Estimate the slope of the line: slope factor
- For cancer use the slope to estimate excessive cancer risk
- Probability that a person will have cancer over a lifetime resulting from exposure 
- Cancer risk = slope factor * Exposure
